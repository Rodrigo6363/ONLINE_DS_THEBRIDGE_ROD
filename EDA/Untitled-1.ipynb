```python
import requests
import pandas as pd

# Consulta a la API de UniProt para obtener información sobre la proteína P42858
url = "https://www.uniprot.org/uniprot/P42858.json"
response = requests.get(url)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    data = response.json()
    
    # Extraer las variantes de la proteína
    variants = data['variants']
    
    # Crear listas para los datos de la tabla
    variant_ids = []
    descriptions = []
    positions = []
    types = []
    cag_lengths = []
    impacts = []
    
    # Iterar sobre las variantes y extraer la información relevante
    for variant in variants:
        variant_ids.append(variant.get('id', 'N/A'))
        descriptions.append(variant.get('description', 'N/A'))
        positions.append(variant.get('position', 'N/A'))
        # Suponiendo que 'type' especifica si es patológica o no
        types.append(variant.get('type', 'N/A'))
        # Si existe un campo para la longitud de la repetición CAG, lo extraemos
        cag_length = variant.get('cag_length', 'N/A')
        cag_lengths.append(cag_length)
        impacts.append(variant.get('impact', 'N/A'))
    
    # Crear un DataFrame de pandas con los datos obtenidos
    df_variants = pd.DataFrame({
        'Variant ID': variant_ids,
        'Description': descriptions,
        'Position': positions,
        'Type': types,
        'CAG Length': cag_lengths,
        'Impact': impacts
    })
    
    # Mostrar la tabla
    print(df_variants)
else:
    print("Error al obtener datos de UniProt")
```
